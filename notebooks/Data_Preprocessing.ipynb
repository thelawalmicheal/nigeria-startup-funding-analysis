```python
df['Investors'] = df['Investors'].fillna('Unknown')
df['Investors Count'] = df['Investors Count'].fillna(0)
```

```python
if 'Effective date' in df.columns:
    df.drop(columns=['Effective date'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Found At'] = pd.to_datetime(df['Found At'], errors='coerce')
```

```python
df['Found At'].unique()
```

```python
median_date = df['Found At'].median()
df['Found At'].fillna(median_date, inplace=True)
```

```python
df['Found At'].unique()
```

```python
df['Year'] = df['Found At'].dt.year
df['Month'] = df['Found At'].dt.month
```

```python
def classify_funding(amount):
    if amount < 10_000_000:
        return 'Small'
    elif amount < 100_000_000:
        return 'Medium'
    else:
        return 'Large'

df['Funding Size'] = df['Amount Normalized'].apply(classify_funding)
```

```python
df['Investors'] = df['Investors'].fillna('Unknown')
df['Investor List'] = df['Investors'].apply(lambda x: x.split(', ') if x != 'Unknown' else [])
df['Calculated Investor Count'] = df['Investor List'].apply(len)
```

```python
df['Investors Count'] = df['Investors Count'].fillna(0)
df['Investor Density'] = df['Amount Normalized'] / (df['Investors Count'] + 1)
```

```python
df['Categories'] = df['Categories'].apply(eval)
df['Category Count'] = df['Categories'].apply(len)
df['Primary Category'] = df['Categories'].apply(lambda x: x[0] if len(x) > 0 else 'Unknown')
```

```python
import ast

def safe_eval(val):
    try:
        return ast.literal_eval(val)
    except:
        return [val] if isinstance(val, str) else []

df['Categories'] = df['Categories'].apply(safe_eval)
```
